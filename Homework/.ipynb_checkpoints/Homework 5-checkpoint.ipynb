```python
import numpy as np
import matplotlib
```

```python
a = np.linspace(-2,2,100)
b = np.linspace(-2,2,100)
x, y = np.meshgrid(a, b) 
```

```python
from matplotlib import pyplot as plt
import math
```

```python
plt.pcolormesh(x,y,y)
```

```python
r = np.sqrt(x**2 + y**2)
p = np.arctan2(y,x)
```

```python
plt.pcolormesh(x,y,r)
```

```python
plt.pcolormesh(x,y,p)
```

```python
f = (np.cos(4*r))**2 + (np.sin(4*p))**2
```

```python
plt.pcolormesh(x,y,f)
```

```python
plt.plot(T, levels)
plt.title("Temperature versus Levels")
plt.xlabel("Temperature in C")
plt.ylabel("Levels")
```

```python
plt.plot(P, levels)
plt.title("Pressure versus Levels")
plt.xlabel("Pressure")
plt.ylabel("Levels")
```

```python
plt.plot(S, levels)
plt.title("Salinity versus Levels")
plt.xlabel("Salinity")
plt.ylabel("Levels")
```

```python
#depth is the row
T_rowmean = np.mean(T, axis=1)
P_rowmean = np.mean(P, axis=1)
S_rowmean = np.mean(S, axis=1)
T_rowsd = np.std(T, axis=1)
P_rowsd = np.std(P, axis=1)
S_rowsd = np.std(S, axis=1)
```

```python
plt.plot(T_rowmean, levels)
plt.title("Average Temperature versus Levels")
plt.xlabel("Average Temperature in C")
plt.ylabel("Levels")
plt.errorbar(T_rowmean,levels, yerr=T_rowsd, ecolor='blue')
```

```python
plt.plot(P_rowmean, levels)
plt.title("Average Pressure versus Levels")
plt.xlabel("Average Pressure")
plt.ylabel("Levels")
plt.errorbar(P_rowmean,levels, yerr=P_rowsd, ecolor='blue')
```

```python
plt.plot(S_rowmean, levels)
plt.title("Average Salinity versus Levels")
plt.xlabel("Average Salinity")
plt.ylabel("Levels")
plt.errorbar(S_rowmean,levels, yerr=S_rowsd, ecolor='blue')
```

```python
#recaculate 
nT_rowmean = np.nanmean(T, axis=1)
nP_rowmean = np.nanmean(P, axis=1)
nS_rowmean = np.nanmean(S, axis=1)
nT_rowsd = np.nanstd(T, axis=1)
nP_rowsd = np.nanstd(P, axis=1)
nS_rowsd = np.nanstd(S, axis=1)
```

```python
plt.plot(nT_rowmean, levels)
plt.title("Average Temperature versus Levels")
plt.xlabel("Average Temperature in C")
plt.ylabel("Levels")
plt.errorbar(nT_rowmean,levels, yerr=nT_rowsd, ecolor='blue')
```

```python
plt.plot(nP_rowmean, levels)
plt.title("Average Pressure versus Levels")
plt.xlabel("Average Pressure")
plt.ylabel("Levels")
plt.errorbar(nP_rowmean,levels, yerr=nP_rowsd, ecolor='blue')
```

```python
plt.plot(nS_rowmean, levels)
plt.title("Average Salinity versus Levels")
plt.xlabel("Average Salinity")
plt.ylabel("Levels")
plt.errorbar(nS_rowmean,levels, yerr=nS_rowsd, ecolor='blue')
```
