## Matplotlib

```python
from matplotlib import pyplot as plt
```

```python
plt.plot(x)
```

```python
f = np.sin(xx)*np.cos(yy)
f
```

```python
plt.pcolormesh(x,y,f, vmin =-1, vmax =1,cmap = 'jet')
plt.colorbar()
```

```python
plt.contourf(x,y,f)
plt.colorbar()
```
