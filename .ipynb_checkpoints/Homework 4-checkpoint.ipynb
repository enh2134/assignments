```python
def KtoC(n):
    """ Convert Kelvin to Celcius """
    C = [n-272.15]
    return C
KtoC(300)
```

```python
def CtoK(n):
    """ Convert Celcius to Kelvin """
    K = [n+272.15]
    return K
CtoK(300)
```

```python
def toF(n, Kelvin = False):
    """ Convert either C or K to Fahrenheit """
    if Kelvin:
        c = n-272.15
        Fahrenheit = (c*(9/5)+32)
    else:
        Celcius = n
        Fahrenheit = (float(Celcius)*(9/5))+32
    return Fahrenheit
```

```python
toF(300, Kelvin = False)
```

```python
toF(300, Kelvin = True)
```

```python
def fromF(f, Kelvin = False):
    """ Convert to C or K from F """
    if Kelvin:
        c = (f-32)*(5/9)
        result = CtoK(c)
    else: 
        result = (f-32)*(5/9)
    return result
```

```python
fromF(300, Kelvin = True)
```

```python
fromF(300, Kelvin = False)
```

```python
def FttoM(f, i):
    b = i*0.0254
    a = f*0.3048
    sum = a+b
    return sum
```

```python
FttoM(5,6)
```

```python
import math
def MtoFI(m):
    c = math.floor(m/0.3048)
    d = round(((m/0.3048)-c)*12)
    return c, d
MtoFI(.5)
```
