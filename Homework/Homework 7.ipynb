```python
import numpy as np
import pandas as pd
import matplotlib 
from matplotlib import pyplot as plt
```

```python
quakes = pd.read_csv(
    'http://www.ldeo.columbia.edu/~danielmw/usgs_earthquakes_2014.csv',
    index_col='id',
    parse_dates=['time', 'updated']
)
quakes
```

```python
quakes.describe
```

```python
top20 = quakes['mag'].nlargest(20)
top20
```

```python
Country = quakes['place']
Country
```

```python
df = pd.DataFrame(top20)
```

```python
df2 = df.join(Country)
df2
```

```python
df2.columns = ['Magnitude', 'Country']
df2
```

```python
df2['Country'].unique()
```

```python
fourup = quakes[quakes['mag'] >= 4]
fourup
```

```python
country_counts = fourup['place'].value_counts()
top5 = country_counts.head(5)

top5.plot(kind='bar', title='Top 5 Locations by Earthquake Count')
plt.xlabel('Country/State')
plt.ylabel('Number of Earthquakes')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

quakes['mag'].plot(kind='hist', ax=axes[0], bins=30)
axes[0].set_title('All Earthquakes')
axes[0].set_xlabel('Magnitude')
axes[0].set_ylabel('Count (log scale)')

fourup['mag'].plot(kind='hist', ax=axes[1], bins=30)
axes[1].set_title('Filtered Earthquakes (mag ≥ 4)')
axes[1].set_xlabel('Magnitude')
axes[1].set_ylabel('Count (log scale)')
```

The log scale shows the magnitudes much closer to the lower end of the scale and the counts have shifted drastically to show more extreme values. 

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

scatter1 = axes[0].scatter(
    quakes['longitude'], quakes['latitude'],
    c=quakes['mag'], cmap='seismic', s=5
)
axes[0].set_title('All Earthquakes')
axes[0].set_xlabel('Longitude')
axes[0].set_ylabel('Latitude')
plt.colorbar(scatter1, ax=axes[0], label='Magnitude')

scatter2 = axes[1].scatter(
    fourup['longitude'], fourup['latitude'],
    c=fourup['mag'], cmap='seismic', s=5
)
axes[1].set_title('Filtered Earthquakes (mag ≥ 4)')
axes[1].set_xlabel('Longitude')
axes[1].set_ylabel('Latitude')
plt.colorbar(scatter2, ax=axes[1], label='Magnitude')
```

The filtered has a much different range of earthquakes as seen by the colors ranges.
